## Dataloaders

```python
transform=Compose([ToTensor()])

train_dataset=MNIST('mnist_train', train=True, download=True,
                             transform=transform)
test_dataset=MNIST('mnist_test', train=False, download=True,
                             transform=transform)
```

```python
train_dataloader=DataLoader(train_dataset,batch_size=BS,shuffle=True)
test_dataloader=DataLoader(test_dataset,batch_size=BS,shuffle=True)
```

## Two Step Clustering comparison

```python

with torch.no_grad():

    n_semi=2000



    stack=[]
    labels=[]
    z_list=[]
    for i,data in enumerate(train_dataset):
        x,y=data
        stack.append(x)
        labels.append(y)
        if i==n_semi-1:
            break


    stack=torch.stack(stack).to(device)
    labels=torch.Tensor(labels)

    z_mean,z_log_var = model.encoder(stack)

    model.q_z = D.normal.Normal(z_mean, torch.exp(0.5 * z_log_var))

    device=z_mean.get_device()
    z = model.q_z.rsample()
    labels=torch.Tensor(labels)
    z_list.append(z)

    #mappings,accuracy=mapping(cl,labels)
```

```python
z_list=z_list[0].cpu().numpy()
labels=labels.cpu()
```

```python
def mapping(cl,y,n_basis):
    """map[k]=label so we map the cluster k-th to the label"""
    
    map={}
    
    acc={}
    
    for k in range(n_basis):
        if len(y[cl==k]):
            v=y[cl==k].mode().values
            map[k]=v
            acc[k]=(y[cl==k]==v).sum()/len(y[cl==k]) 

    return map,acc
```

```python
with torch.no_grad():
    recon=[]
    y_pred=[]
    test_labels=[]
    z_sampled=[]
    for x,y in tqdm.tqdm(test_dataset):
        
        
        z_mean,z_log_var = model.encoder(x.unsqueeze(0).to(device))

        model.q_z = D.normal.Normal(z_mean, torch.exp(0.5 * z_log_var))

        device=z_mean.get_device()
        z = model.q_z.rsample()
        
        
        #y_pred.append(int(mappings[cluster].numpy()))
        test_labels.append(y)
        z_sampled.append(z.detach().cpu().numpy())

    z_sampled=np.array(z_sampled)
    
```

```python
def tsne_vis(data,kdir,labels=test_labels):
    
    tsne=TSNE(n_components=2, learning_rate='auto',init='random').fit_transform(np.squeeze(data))
    cdict = {0: 'red', 1: 'blue', 2: 'green',3: 'yellow', 4: 'orange', 5: 'pink', 6:'tab:blue', 7:'purple',8:'brown',9:'gray'}

    fig, ax = plt.subplots()
    for g in np.unique(labels):
        ix = np.where(labels == g)
        ax.scatter(tsne[:,0][ix], tsne[:,1][ix], c = cdict[g], label = g, s = 20,alpha=0.7,linewidth=1.)
    #plt.scatter(tsne[:,0],tsne[:,1],label=labels,c= cdict[g])
    ax.legend()
    ax.set_title("tSNE represantation of the latent space")
    plt.savefig(os.path.join(kdir,"tsne.png"))
    
    
```

```python
def evaluate_cluster_performances(noc,z_list=z_list,z_sampled=z_sampled,labels=labels,test_labels=test_labels,mode="km"):
    
    print(f"Clustering with {noc} cluster")
    if mode=="km":
        kmodel = KMeans(noc)
    kmodel.fit(z_list)

    mappings,accuracy=mapping(kmodel.predict(z_list),labels,n_basis=noc)
    clusters=kmodel.predict(z_sampled.squeeze())
    y_pred=[mappings[c] for c in clusters]
    
    kdir=os.path.join(base_path,f"{mode}_{noc}")
    os.makedirs(kdir,exist_ok=True)
    f = open(os.path.join(kdir,"classification_report.txt"), "w")
    f.write(classification_report(test_labels, y_pred))
    f.close()
    
    prototypes=kmodel.cluster_centers_
    x2 = model.decoder(torch.Tensor(prototypes).to(device))
    
    fig, axs= plt.subplots(1,noc,figsize=(20,10))

    for i,ax in enumerate(axs.ravel()):
        ax.imshow(x2[i].cpu().detach().permute(1,2,0).numpy())

    plt.savefig(os.path.join(kdir,"prototypes.png"))
    
    print(f"Computing tsne representations")
    tsne_vis(z_sampled,kdir)
    
```

```python

for noc in tqdm.tqdm([15,20,25,30,35,40,50]): 
    evaluate_cluster_performances(noc)
```
